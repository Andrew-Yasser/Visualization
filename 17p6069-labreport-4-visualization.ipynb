```python
#Importing Dataset
df = pd.read_csv("../input/top-play-store-games/android-games.csv")
```

```python
#Checking for NaN values
df.isna().sum()
#Removing NaN Values
df.dropna(inplace = True)
```

```python
#Creating the text variable
text = " ".join(cat.split()[1] for cat in df.category)
```

```python
# Creating word_cloud with text as argument in .generate() method
Word_Cloud = WordCloud(collocations = False, background_color = 'white').generate(text)
```

```python
# Display the generated Word Cloud
plt.imshow(Word_Cloud, interpolation='bilinear')
plt.axis("off")
plt.title("this is my word cloud using python")
plt.show()
```

```python
X = np.random.rand(10, 12)
categories = ['Card', 'Arcade', 'Puzzle', 'Racing', 'Adventure', 'Simulation', 'Action', 'Role', 'Sports', 'Music']
#Display the generated dendrogram
fig = ff.create_dendrogram(X, orientation='left', labels=categories)
fig.update_layout(width=1000, height=1000)
fig.show()
```
